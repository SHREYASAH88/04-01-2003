```python
def plot_hist(variable):
    plt.figure(figsize = (9,5))
    plt.hist(df[variable], bins = 10)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distrubiton with hist".format(variable))
    plt.show()
```
